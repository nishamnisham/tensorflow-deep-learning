```python
# A classification problem practice
# The notebook is made for quick references

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import tensorflow as tf
from sklearn.datasets import make_circles
import matplotlib.pyplot as plt
```

```python
# create circles
# Number of circle points to create
n_samples = 1000

# Creating circles
X, y = make_circles(n_samples,
                   noise = 0.04,
                   random_state = 42)
```

```python
# What are the shapes
X.shape,y.shape
```

```python
# Visualise
# Create a dataframe to make visualisation easy

df = pd.DataFrame({'X1':X[:,0],'X2':X[:,1], 'labels': y})
```

```python
df.head()
```

```python
# Visualise with a plot
plt.scatter(df['X1'],df['X2'], c=y, cmap=plt.cm.RdYlBu)
```

# Creating a model

```python
# 1. Create the model
# The model will be sequential

model = tf.keras.Sequential([
    tf.keras.layers.Dense(100), # Dense neurones
    tf.keras.layers.Dense(100), # Dense neurones
    tf.keras.layers.Dense(100), # Dense neurones
    tf.keras.layers.Dense(1)    # Output layers
])


# 2. Compile the model
model.compile(loss=tf.keras.losses.BinaryCrossentropy(),
             optimizer=tf.keras.optimizers.Adam(),
             metrics=['accuracy'])

# loss function is Binary Cross entropy because this is a binary problem where it predicts blue/red (1/0)
```

```python
# 3. Fit the model

model.fit(X,y, batch_size=20, epochs=100, verbose=1)
```

```python
# 4. Evaluate the model
model.evaluate(X,y)
```
